```python
# Download and read the `spacex_launch_geo.csv`
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/spacex_launch_geo.csv'
resp = await fetch(URL)
spacex_csv_file = io.BytesIO((await resp.arrayBuffer()).to_py())
spacex_df=pd.read_csv(spacex_csv_file)
spacex_df.head()
```

```python
# Select relevant sub-columns: `Launch Site`, `Lat(Latitude)`, `Long(Longitude)`, `class`
spacex_df = spacex_df[['Launch Site', 'Lat', 'Long', 'class']]
launch_sites_df = spacex_df.groupby(['Launch Site'], as_index=False).first()
launch_sites_df = launch_sites_df[['Launch Site', 'Lat', 'Long']]
launch_sites_df
```

```python
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)
site_map
```

```python
## Task 1: Mark all launch sites on a map
for i, row in launch_sites_df.iterrows():
    coord = [row['Lat'], row['Long']]
    
    # Add Circle
    circle = folium.Circle(
        location=coord,
        radius=1000,
        color="#d35400",
        fill=True
    )
    site_map.add_child(circle)

    # Add Label
    label = folium.map.Marker(
        coord,
        icon=DivIcon(
            icon_size=(20,20),
            icon_anchor=(0,0),
            html=f'<div style="font-size:12px; color:#d35400;"><b>{row["Launch Site"]}</b></div>'
        )
    )
    site_map.add_child(label)

site_map

```

```python
# TASK 3: Calculate the distances between a launch site to its proximities
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

1. Are launch sites in close proximity to railways?

Yes. Each SpaceX launch site has a railway or railway spur very close to the launch pad.

2. Are launch sites in close proximity to highways?

Yes. All launch sites are located near major roads or highways.
Highways provide ground access for personnel, emergency vehicles, and transportation of materials.



3. Are launch sites in close proximity to coastline?

Yes. All launch sites are extremely close to the coastline.
This is intentional, launching over the ocean greatly reduces risk to populated areas in the event of an accident during ascent.
It also allows a safer flight path for reusable boosters returning to land or drone ships.



4. Do launch sites keep certain distance away from major cities?

Yes. Launch sites are typically located far from major cities.
This distance reduces:
		risk to large populations
		noise impact
		airspace disruption

They are close enough for logistical support, but far enough to minimize danger.
