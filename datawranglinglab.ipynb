```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv("dataset_part_1.csv")
df.head()
```

```
df.info()
```

```
df.isnull().sum()
```

```
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```
payload_mean = df['PayloadMass'].mean()
df['PayloadMass'] = df['PayloadMass'].fillna(payload_mean)
```

```
df['LandingPad'] = df['LandingPad'].fillna("Unknown")
df['Serial'] = df['Serial'].fillna("Unknown")
df['Block'] = df['Block'].fillna(-1)
df['ReusedCount'] = df['ReusedCount'].fillna(0)
```

```
df['FlightNumber'] = df['FlightNumber'].astype(int)
df['Date'] = pd.to_datetime(df['Date'])
numeric_cols = ['PayloadMass', 'Flights', 'Block', 'ReusedCount', 'Longitude', 'Latitude']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric)
```

```
df['Year'] = df['Date'].dt.year
df['LandingSuccess'] = df['Outcome'].apply(lambda x: 1 if "True" in x else 0)
```

```
df.info()
df.head()
```

```
df.to_csv("dataset_part_2.csv", index=False)
```
